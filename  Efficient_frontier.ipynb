```python
from pandas_datareader import data as pdr
import numpy as np
import datetime as dt
import pandas as pd
import yfinance as yf
```

```python
start= dt.datetime(2010,1,1)
end=dt.datetime.now()
```

```python
# Extract Nifty 50 companies ticker name
data1=pd.read_csv("Ticker.csv")
ticker_data = data1['Company'].tolist()
ticker_data
```

```python
yf.pdr_override()
#tickers=['RELIANCE.NS','KOTAKBANK.NS','SBIN.NS','HDFCBANK.NS','INFY.NS','ITC.NS','BHARTIARTL.NS']
tickers=ticker_data

data_df=pd.DataFrame()

for ticker in tickers:
    df=pdr.get_data_yahoo(ticker,start,end)
    data_df[ticker]=df['Adj Close']
```

```python
data_df.head()
```

```python
#Daily Log return ( Additive return )
return1=np.log(data_df/data_df.shift(1)) # Gives log(priice[t]/price[t-1])
return1
```

```python
#Cleaning data " return1 "
return1.dropna(inplace=True)
return1.head()
```

```python
# Getting Variance and Standard deviation from data description
# Efficient frontier has two variable: Expected Return(Mean Return) and Risk(Standard Deviation)
return1.describe()
```

```python
er=np.array(return1.describe().loc['mean'])
std=np.array(return1.describe().loc['std'])
```

```python
sharpe=(return1.describe().loc['mean'])/(return1.describe().loc['std'])
sharpe
```

```python
# Computation of Variance and Covariance matrix 
varcov=return1.cov()
varcov
# The diagonal numbers represent the variance 
```

```python
import seaborn as sns 
plt.figure(figsize=(14, 14))
dataplot=sns.heatmap(varcov) 
plt.show()
```

```python
# Efficient Frontier requires multiple portfolios to analyse the optimum portfolio
# We consider two portfolios for this case
```

```python
# Efficient Portfolio A

c1=0

weightA=np.linalg.inv(varcov) @ (er-c1)/np.sum(np.linalg.inv(varcov) @ (er-c1))
erA= weightA@er
varA=weightA@ varcov @ weightA.T
stdA=np.sqrt(varA)
sharpeA=erA/stdA
```

```python
sharpeA
```

```python
# Efficient Portfolio B

c2=0.04

weightB=np.linalg.inv(varcov) @ (er-c2)/np.sum(np.linalg.inv(varcov) @ (er-c2))
erB= weightB@er
varB=weightB@ varcov @ weightB.T
stdB=np.sqrt(varB)
sharpeB=erB/stdB
```

```python
sharpeB
```

```python
# Covariance of Portfolio A and B
covAB=weightA@varcov@weightB.T
```

```python
# Creating Random weights for Portfolio A for using in C 
# We use linespace function to create linear differenced list of weight inbetween a range
wA=np.linspace(-0.5,4.5,num=1000)
wB=1-wA
```

```python
# Computing Expected Return of Portfolio C ( Portfolio A + B)
erC=wA*erA + wB*erB
```

```python
# Computing Standard deviation of portfolio C
stdC=np.sqrt( (wA*stdA)**2 + (wB*stdB)**2 + 2* wA*wB*covAB)
```

```python
# Calculate Global minimum variance portfolio

# Unit Matrix
UM=np.ones((varcov.shape[0]))

GMVP= UM @ np.linalg.inv(varcov)/(UM @ np.linalg.inv(varcov)@UM.T)
```

```python
erG= GMVP@er
varG=GMVP@ varcov @ GMVP.T
stdG=np.sqrt(varG)
```

```python
# Plot the Efficient Frontier

import matplotlib.pyplot as plt

plt.plot(stdC,erC)
plt.scatter(stdG, erG)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(
    stdC,
    erC,
    linestyle="--",
    color="blue",
    label="Efficient Frontier",
    linewidth=2,
    alpha=0.6,)

plt.scatter(stdG, erG, color="green", label="GMVP")
plt.annotate(
    f"MVP\nRisk: {stdG*100:.2f}%\nReturn: {erG*100:.2f}%",
    (stdG, erG),
    textcoords="offset points",
    xytext=(-30, 10),)
```
